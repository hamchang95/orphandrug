```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
orphandrug = {}
list_drug = []
list_manuf = []

for pagenumber in range (1,4):
    url = 'http://apis.data.go.kr/1470000/RareMdcinInfoService/getRareMdcinList?ServiceKey=0wCzOGsG4fb2lPsmgwt0SFagZ2WY6uf2IYClQV%2FccJX5F1%2FB93kJydwtFS4DyLunJIHYeqSP6hDqH9Rsuhw3Uw%3D%3D&numOfRows=100&pageNo='+ str(pagenumber)
    req = requests.get(url)
    html = req.text
    soup = BeautifulSoup(html, 'html.parser')
    drug = soup.find_all('goods_name')
    manuf = soup.find_all('manuf_name')
    for i in drug:
        list_drug.append(i.text)
    for x in manuf:
        list_manuf.append(x.text)


orphandrug['Orphan Drug'] = list_drug
orphandrug['Manufacturer'] = list_manuf

df = pd.DataFrame(orphandrug)
df
```
